```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
data=pd.read_csv('Downloads/myexcel - myexcel.csv (1).csv')
data
```

```python
data.isnull().sum()
```

```python
x=data['Salary'].mean()
data['Salary'].fillna(x,inplace = True)
data
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Height']=np.random.uniform(150,180,size=len(data))
```

```python
data
```

**1:How many are there in each team and percentage splitting with respect to the total employees**

```python
data['Team'].value_counts()
```

**2:Percentage splitting with respect to the total employees**

```python
data['Team'].value_counts()/len(data)*100
```

**3:Segregate the employees with respect to different positions**

```python
employees = data.groupby('Position')['Name'].apply(list)
for Position, Names in employees.items():
    print(f"employees in {Position} position:")
    for name in Names:
        print(name)
        print("\n")
```

**4:Find from which age group most of the employees belong to**

```python
age_count=data['Age'].value_counts()
age_count
```

**5:Find out under which team and position,spending in terms of salary is high**

```python
spending_salary=data.groupby(['Team','Position'])['Salary'].sum()
spending_salary.idxmax()
```

Find if there is any correlation btw age and salary,represent it visually

```python
correlation=data['Salary'].corr(data['Age'])
```

```python
print("THE CORRELATION B/w SALARY AND AGE IS:",correlation)
```

```python
sns.scatterplot(x="Age",y="Salary",data=data)
plt.ylabel("Salary")
plt.xlabel("Age")
plt.title("Correlation btw Salary and Age")
plt.show()
```
